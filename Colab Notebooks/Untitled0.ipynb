```python
import random
import matplotlib.pyplot as plt
NO_OF_CHARS = 256
import time
letters = 'abcd'
def badcharatable(pattern,badChar):

    # Fill the actual value of last occurrence
    for i in range(len(pattern)):
        badChar[ord(pattern[i])] = i

    # return initialized list
    return badChar

def find(text, pattern,b):
    n = len(text)
    m = len(pattern)
    badChar = badcharatable(pattern,b)
    # s is shift of the pattern with respect to text
    s = 0
    while(s <= n-m):
        j = m-1

        # Keep reducing index j of pattern while
        # characters of pattern and text are matching
        # at this shift s
        while j>=0 and pattern[j] == text[s+j]:
            j -= 1

        # If the pattern is present at current shift,
        # then index j will become -1 after the above loop
        if j<0:
            print("Pattern shift at = {}".format(s))
            s += (m-badChar[ord(text[s+m])] if s+m<n else 1)
        else:
            s += max(1, j-badChar[ord(text[s+j])])


def boyer():
    for i in inputs:
        badChar = [-1]*NO_OF_CHARS
        text=i[0]
        pattern=i[1]
        start = time.time()
        find(text, pattern,badChar)
        end = time.time()
        diff = end - start
        timer.append(diff)


inputSizes =[]
inputs = []
# random.seed(10)

for i in range(500,1500):
    text=""
    pattern=""
    inputSizes.append(i)
    for j in range(i):
        a = random.randint(0,3)
        text+=(letters[a])


    for j in range(5):
        a = random.randint(0,3)
        pattern+=(letters[a])
    inputs.append((text,pattern))

timer=[]
boyer()
print(timer)
plt.figure()
plt.scatter(inputSizes,timer, color='blue', marker='o', label='Scatter Plot')
# plt.plot(inputSizes,timer, label='boyer moore' , color = 'red')
plt.legend(loc="upper left")

plt.xlabel('Number of inputs')

plt.ylabel('Time Taken')
plt.title('Boyer-Moore algorithm')

plt.show()
```
