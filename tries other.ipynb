```python
from chemprop.data import MoleculeDatapoint, MoleculeDataset

def generate_rdkit_features(mol):
    if mol is None:
        raise ValueError("Invalid molecule")
    features = [
        Descriptors.MolWt(mol),  # Molecular weight
        Descriptors.NumHDonors(mol),  # Number of hydrogen bond donors
        Descriptors.NumHAcceptors(mol),  # Number of hydrogen bond acceptors
        Descriptors.NumRotatableBonds(mol),  # Number of rotatable bonds
        Descriptors.TPSA(mol)  # Topological polar surface area
        # Add more features as needed
    ]
    return features

def apply_featurizer(data, generate_features):
    featurized_data = []
    for datapoint in data:
        mol = Chem.MolFromSmiles(datapoint.smiles)
        features = generate_features(mol)  # Generate features using the function
        # Create MoleculeDatapoint with smiles, features, and targets
        featurized_datapoint = MoleculeDatapoint(smiles=datapoint.smiles, features=features, targets=datapoint.targets)
        featurized_data.append(featurized_datapoint)
    return featurized_data

# Apply the feature generation function to all datasets
train_data_featurized = apply_featurizer(train_data, generate_rdkit_features)
val_data_featurized = apply_featurizer(val_data, generate_rdkit_features)
test_data_featurized = apply_featurizer(test_data, generate_rdkit_features)

# Create MoleculeDataset objects with the featurized data
train_dset = MoleculeDataset(train_data_featurized)
val_dset = MoleculeDataset(val_data_featurized)
test_dset = MoleculeDataset(test_data_featurized)

# Normalize targets
scaler = train_dset.normalize_targets()

# Normalize targets for validation and test datasets using the scaler
val_targets = [datapoint.targets for datapoint in val_dset]
test_targets = [datapoint.targets for datapoint in test_dset]

# Apply normalization
val_targets_normalized = scaler.transform(val_targets)
test_targets_normalized = scaler.transform(test_targets)

# Update datasets with normalized targets
for datapoint, target in zip(val_dset, val_targets_normalized):
    datapoint.targets = target

for datapoint, target in zip(test_dset, test_targets_normalized):
    datapoint.targets = target
```
