```python
import os
import zipfile

# CREATE A DIRECTORY FOR MOLFINGERPRINTS
os.makedirs("/content/drive/MyDrive/BBB dataset/Data_Regression/MolFingerprints_Rg", exist_ok=True)

# DOWNLOAD THE FINGERPRINTS_XML.ZIP FILE
os.system('curl -LJO https://github.com/dataprofessor/padel/raw/main/fingerprints_xml.zip')

# UNZIP THE DOWNLOADED FILE
with zipfile.ZipFile('fingerprints_xml.zip', 'r') as zip_ref:
    zip_ref.extractall('/content/drive/MyDrive/BBB dataset/Data_Regression/MolFingerprints_Rg')
```

```python
# LIST AND SORT FINGERPRINT XML FILES
xml_files = glob.glob("/content/drive/MyDrive/BBB dataset/Data_Regression/MolFingerprints_Rg/*.xml")
xml_files.sort()
```

```python
# DELETE THE MOST COMPUTATIONAL DEMANDING MOLECULAR FINGERPRINTS
xml_files.remove("/content/drive/MyDrive/BBB dataset/Data_Regression/MolFingerprints_Rg/KlekotaRothFingerprintCount.xml")
xml_files.remove("/content/drive/MyDrive/BBB dataset/Data_Regression/MolFingerprints_Rg/KlekotaRothFingerprinter.xml")
xml_files
```

```python
# CREATE A LIST WITH SHORTEN NAMES OF XML FILES
FP_list = ['AtomPairs2DCount',
 'AtomPairs2D',
 'EState',
 'CDKextended',
 'CDK',
 'CDKgraphonly',
 'MACCS',
 'PubChem',
 'SubstructureCount',
 'Substructure']
```

```python
# CREATE A DICTIONARY WITH SHORTEN NAMES AS KEYS AND XML FILE NAMES AS VALUES
fp = dict(zip(FP_list, xml_files))
fp
```

```python
# PREPARE DATA SUBSET AS INPUT TO PADELPY AND EXPORT IT AS .SMI FILE
#FOR pIC50
df_padelpy1 = df['SMILES']
df_padelpy1.to_csv('/content/drive/MyDrive/BBB dataset/Data_Regression/molecule_pIC50.smi', sep='\t', index=False, header=False)
df_padelpy1
```

```python
import os

for i in fp:
    path = "/content/drive/MyDrive/BBB dataset/Data_Regression/MolFingerprints_Rg/Results_pIC50/Original_pIC50/"
    fingerprint = i
    fingerprint_output_file_path = ''.join([path, fingerprint])
    fingerprint_output_file = ''.join([fingerprint_output_file_path, '.csv'])
    fingerprint_descriptortypes = fp[fingerprint]  # This variable holds the path to the xml file

    # Print the value of fingerprint_descriptortypes to verify it is correct
    print(fingerprint_descriptortypes)

    # Check if the descriptor types file exists
    if not os.path.exists(fingerprint_descriptortypes):
        raise FileNotFoundError(f"Descriptor types file not found: {fingerprint_descriptortypes}")

    if not os.path.exists(path):
        raise FileNotFoundError(f"path types file not found: {path}")

    padeldescriptor(mol_dir='/content/drive/MyDrive/BBB dataset/Data_Regression/molecule_pIC50.smi',
                    d_file=fingerprint_output_file,
                    descriptortypes=fingerprint_descriptortypes,
                    detectaromaticity=True,
                    standardizenitro=True,
                    standardizetautomers=True,
                    threads=2,
                    removesalt=True,
                    log=True,
                    fingerprints=True)
```
